## Merging both Dataframes together

```python
df = movies.merge(directors,
                  left_on = "director_id",
                  right_on = "id",
                  how = "left",
                  suffixes = ["_movies","_directors"]
                  )
df
```

## Dropping unnecessary columns

```python
df.drop(columns = ["director_id","id_directors"], inplace = True)

# Since we have director's names, we don't need their IDs
```

```python
df.head()
```

## Renaming the column

```python
df.rename(columns = {"id_movies":"movies_id"}, inplace = True)
```

```python
df.head()
```

## Converting budget and revenue into millions

```python
df["budget"] = round(df["budget"]/1000000,2)
```

```python
df["revenue"] = round(df["revenue"]/1000000,2)
```

```python
df.head()
```

```python
df.rename(columns = {"budget":"budget_mil","revenue":"revenue_mil"}, inplace = True)
```

```python
df.head() # Now we have a nice and clean data for analysis
```

## Calculate the career span and productivity rate for the same director


Career span: The total number of years they have devoted to the industry.

Productivity Rate: Calculated as the total number of movies released by the director divided by career span.

```python
df_jc = df.loc[df["director_name"] == "James Cameron"]
df_jc
```

```python
df_jc["year"].max()
```

```python
df_jc["year"].min()
```

```python
career_span = df_jc["year"].max() - df_jc["year"].min()
career_span
```

```python
df_jc["title"].nunique()
```

```python
productivity = df_jc["title"].nunique()/career_span
productivity
```

* **James Cameron** has a **25-year** career span with **0.28** productivity rate.

## Calculate career span and productivity rate for all the directors

```python
df.groupby(["director_name"]).ngroups # Total no. of unique directors present in the dataframe
```

```python
df.groupby(["director_name"]).groups # All the directors with their set of index
```

```python
directors_productivity = df.groupby(["director_name"]).agg(year_max = ("year","max"),
                                    year_min = ("year", "min"),title_count = ("title","count")).reset_index()
directors_productivity
```

```python
directors_productivity["career_span"] = directors_productivity["year_max"] - directors_productivity["year_min"]
directors_productivity["productivity_rate"] = directors_productivity["title_count"]/directors_productivity["career_span"]
```

```python
directors_productivity
```

```python
directors_productivity.sort_values(by = "productivity_rate", ascending = False)

# High productivity_rate directors
```

## Calculate all the risky movies

risky movies = budget > revenue

```python
def calculate_risky_movies(x):
  x["risky"] = (x["budget_mil"]-x["revenue_mil"].mean())>=0
  return x
```

```python
risky_movies = df.groupby(["director_name"]).apply(calculate_risky_movies)
```
